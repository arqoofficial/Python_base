## Задача

Вам дан массив данных, представленный в виде списка списков. Каждый вложенный список представляет собой измерение одного параметра для разных индивидуумов. В процессе обработки данных вам часто придется сталкиваться с тем, что параметры могут иметь разную шкалу.

Существуют различные способы решения этой проблемы, среди которых — т. н. minmax преобразование, приводящее значения к шкале от 0 до 1.

Формула выглядит следующим образом (в данном случае x — каждое значение, т. е. операцию ниже необходимо выполнить с каждым значением наших данных. Под минимальным и максимальным значениями имеются ввиду минимум и максимум в каждом вложенном списке, а не во всех данных):

**X(scaled) = [X - X(min)]/[X(max) - X(min)]**

Вам необходимо создать функцию, которая принимает на вход массив, в качестве результата получить такой же список списков, но при этом наполняющие его значения должны быть преобразованы согласно формуле.

*ВАЖНО!*
*Помните, что любую сколько угодно сложную задачу можно (и нужно) делить на простые составляющие.*

ПОДСКАЗКА
Вам могут помочь встроенные функции min() и max().

## Код:

```python
# Input:
data = [
    [25, -19, 35, 90, 73],
    [14.86452745, -26.8915466, -4.42219305, 6.3526264, 7.88297253],
    [0, 0, 1, 0, 1],
    [837, 302, 790, 271, 796],
    [49.97, 49.94, 51.37, 50.97, 51.29]]

# Solution:
res = []
for data_list in data:
    x_max = max(data_list)
    x_min = min(data_list)
    data_list_scaled = []
    for x in data_list:
        x_scaled = (x-x_min) / (x_max-x_min)
        data_list_scaled.append(x_scaled)
        
    res.append(data_list_scaled)
    print(data_list_scaled)
print()
# Output:
print(res)

        
```

```python
# Function
def scale(list_of_lists):
    res = []
    for data_list in list_of_lists:
        x_max = max(data_list)
        x_min = min(data_list)
        data_list_scaled = []
        for x in data_list:
            x_scaled = (x-x_min) / (x_max-x_min)
            data_list_scaled.append(x_scaled)  
        res.append(data_list_scaled)
    return res

# Input:
data = [
    [25, -19, 35, 90, 73],
    [14.86452745, -26.8915466, -4.42219305, 6.3526264, 7.88297253],
    [0, 0, 1, 0, 1],
    [837, 302, 790, 271, 796],
    [49.97, 49.94, 51.37, 50.97, 51.29]]

# Output:
print(scale(data))
```
