### TRAIN.CSV
1.	row_id: (int64) ID code for the row.
2.	timestamp: (int64) the time in milliseconds between this user interaction and the first event completion from that user.
3.	user_id: (int32) ID code for the user.
4.	content_id: (int16) ID code for the user interaction
5.	content_type_id: (int8) 0 if the event was a question being posed to the user, 1 if the event was the user watching a lecture.
6.	task_container_id: (int16) Id code for the batch of questions or lectures. For example, a user might see three questions in a row before seeing the explanations for any of them. Those three would all share a task_container_id.
7.	user_answer: (int8) the user's answer to the question, if any. Read -1 as null, for lectures.
8.	answered_correctly: (int8) if the user responded correctly. Read -1 as null, for lectures.
9.	prior_question_elapsed_time: (float32) The average time in milliseconds it took a user to answer each question in the previous question bundle, ignoring any lectures in between. 10.	prior_question_had_explanation: (bool) Whether or not the user saw an explanation and the correct response(s) after answering the previous question bundle, ignoring any lectures in between. The value is shared across a single question bundle, and is null for a user's first question bundle or lecture. Typically the first several questions a user sees were part of an onboarding diagnostic test where they did not get any feedback.
 
### QUESTIONS.CSV: METADATA FOR THE QUESTIONS POSED TO USERS.
1.	question_id: foreign key for the train/test content_id column, when the content type is question (0).
2.	bundle_id: code for which questions are served together.
3.	correct_answer: the answer to the question. Can be compared with the train user_answer column to check if the user was right.
4.	part: the relevant section of the TOEIC test.
5.	tags: one or more detailed tag codes for the question. The meaning of the tags will not be provided, but these codes are sufficient for clustering the questions together.
 
### LECTURES.CSV: METADATA FOR THE LECTURES WATCHED BY USERS AS THEY PROGRESS IN THEIR EDUCATION.
1.	lecture_id: foreign key for the train/test content_id column, when the content type is lecture (1).
2.	part: top level category code for the lecture.
3.	tag: one tag codes for the lecture. The meaning of the tags will not be provided, but these codes are sufficient for clustering the lectures together.
4.	type_of: brief description of the core purpose of the lecture.

### 3.1.1. timestamp
Этот столбец характеризует время в милисекундах между данным взаимодействием пользователя и первым завершением события от этого пользователя. То есть это время от начала работы данного пользователя с платформой до данного взаимодействия с конкретным **'content_id'**.

Мы также можем наблюдать, что датасет **train_df** отсортирован сначала по уникальным **'user_id'**, а затем по **'timestamp'**. То есть мы можем проследить динамику ответов конкретного пользователя во времени.

Построим гистограмму по столбцу **'timestamp'**. Для более адекватного восприятия значений времени переведем значения в месяцы. Гугл говорит, что в году 31536000000 милисекунд:

```python
fig = plt.figure(figsize=(15,5))
timestamp = train_df['timestamp']/(31536000000/12)
timestamp.plot.hist(bins=100)
plt.title("Гистограмма столбца timestamp")
plt.xlabel("Число месяцев между данным взаимодействием и первым взаимодействием пользователя")
plt.ylabel("Частота взаимодействий")
plt.show()
```

Видно, что пользователи наиболее активны в самом начале своего взаимодействия с платформой. Чем дольше человек на платформе, тем реже он с ней взаимодействует, что в целом логично.

Для удобства создадим новый признак **'ts_bin'** с помощью квартильного разбиения на 10 квартилей (10 равных по числу взаимодействий категорий):

```python
bins = []
for i in range(1, 11):
    bins.append(f'b{i}')
bins
```

```python
train_df['ts_bin'] = pd.qcut(train_df['timestamp'], q=10, labels=bins)
train_df['ts_bin'].value_counts()
```

Также создадим функцию, которая будет считать процент правильных ответов для конкретного столбца в датафрейме:

```python
def is_correct(col):
    correct = train_df[train_df.answered_correctly != -1].groupby([col, 'answered_correctly'], as_index=False).size()
    correct = correct.pivot(index= col, columns='answered_correctly', values='size')
    correct['Percent_correct'] = round(correct.iloc[:,1]/(correct.iloc[:,0] + correct.iloc[:,1]),2)
    correct = correct.sort_values(by = "Percent_correct", ascending = False)
    correct = correct.iloc[:,2]
    return(correct)
```

```python
bins_correct = is_correct("ts_bin")
bins_correct = bins_correct.sort_index()*100
```

Выведем проценты правильных ответов для каждого квартиля:

```python
bins_correct
```

Видим, что число правильных ответов для первого квартиля несколько ниже, чем для последующих, то есть в самом начале своей работы пользователи делают больше ошибок, а затем постепенно уменьшают процент промахов:

```python
plt.figure(figsize=(12,6))
plt.bar(bins_correct.index, bins_correct.values)
plt.title("Процент правильных ответов в зависимости от квартиля 'timestamp'")
plt.xlabel('Квартиль')
plt.ylabel('Процент')
plt.xticks(rotation=0)
plt.show()
```

Всё отлично, пустых тегов нет. Теперь переведем все теги в списки:

Теперь заведем список тегов (с повторениями) для того, чтобы учесть частоту каждого тега:

```python
q_tags_list = []
for i in questions_df['tags'].values:
    q_tags_list.extend(i)
```

Рассчитаем среднее количество тегов на вопрос:

```python
print(f"Среднее число тегов на вопрос = {round(len(q_tags_list) / questions_df.shape[0], 2)}")
```

Выведем число уникальных тегов:

```python
q_tags_set = set(q_tags_list)
print(f"Число уникальных тегов = {len(q_tags_set)}")
```

Заведем список тегов в объект Series:

```python
q_tags_series = pd.Series(q_tags_list)
q_tags_series.value_counts()
```

Покажем наиболее часто используемые тэги:

Здесь видим явных 4 лидера по тегам: 92, 38, 81 и 29.

Теперь посчитаем число верных и неверных ответов в зависимости от **'question_id'**:

```python
correct_q = train_df[train_df.answered_correctly != -1].groupby(["content_id", 'answered_correctly'], as_index=False).size()
correct_q = correct_q.pivot(index= "content_id", columns='answered_correctly', values='size')
correct_q.columns = ['Wrong', 'Right']
correct_q = correct_q.fillna(0)
correct_q[['Wrong', 'Right']] = correct_q[['Wrong', 'Right']].astype(int)
questions_df = questions_df.merge(correct_q, left_on = "question_id", right_on = "content_id", how = "left")
questions_df.head()
```

Добавим новый столбец с процентом правильных ответов на вопрос с данным **'question_id'**:

```python
questions_df['Percent_of_correct'] = round((questions_df['Right'] / (questions_df['Right'] + questions_df['Wrong']))*100, 1)
questions_df.head()
```

Посмотрим на 50 самых сложжных вопросов, на которые было дано менее 20% правильных ответов:

```python
plt.figure(figsize=(12,5))
questions_df['Percent_of_correct'].sort_values().head(50).plot.bar()
plt.title('50 сложных вопросов')
plt.xlabel('question_id')
plt.ylabel('Процент правильных ответов')
plt.xticks(rotation=90)
plt.show()
```

Самые простые вопросы были отвечены верно почти всеми:

```python
plt.figure(figsize=(12,5))
questions_df['Percent_of_correct'].sort_values().tail(50).plot.bar()
plt.title('50 простых вопросов')
plt.xlabel('question_id')
plt.ylabel('Процент правильных ответов')
plt.xticks(rotation=90)
plt.show()
```

Создадим табличку по тегам. В качестве индексов используем номера тегов. Для каждого тега приведем следующую информацию:
* Номер тега
* Число неправильных и правильных ответов
* Общее число ответов
* Число вопросов с данным тегом
* Процент правильных ответов для данного тега

```python
q_tags_list = list(q_tags_set)
tags_df = pd.DataFrame()
for x in range(len(q_tags_list)):
    df = questions_df[questions_df.tags.apply(lambda l: q_tags_list[x] in l)]
    df1 = df.agg({'Wrong': ['sum'], 'Right': ['sum']})
    df1['Total_questions'] = df1.Wrong + df1.Right
    df1['Question_ids_with_tag'] = len(df)
    df1['tag'] = q_tags_list[x]
    df1 = df1.set_index('tag')
    tags_df = pd.concat([tags_df, df1])

tags_df[['Wrong', 'Right', 'Total_questions']] = tags_df[['Wrong', 'Right', 'Total_questions']].astype(int)
tags_df['Percent_of_correct'] = tags_df.Right/tags_df.Total_questions
tags_df = tags_df.sort_values(by = "Percent_of_correct")
```

```python
plt.figure(figsize=(12,5))
tags_df['Percent_of_correct'].head(15).plot.bar()
plt.title('15 сложных тегов')
plt.xlabel('tag')
plt.ylabel('Процент правильных ответов')
plt.xticks(rotation=0)
plt.show()
```

Выведем 15 самых легких тегов:

```python
plt.figure(figsize=(12,5))
tags_df['Percent_of_correct'].tail(15).plot.bar()
plt.title('15 сложных тегов')
plt.xlabel('tag')
plt.ylabel('Процент правильных ответов')
plt.xticks(rotation=0)
plt.show()
```
